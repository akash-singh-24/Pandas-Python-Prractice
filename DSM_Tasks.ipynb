```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

df_tasks = pd.read_csv('export (1).csv')
df_subtasks = pd.read_csv('export (2).csv')
```

```python
df_tasks1 = df_tasks[['_id','store_id','type','status','creation_type']]
df_tasks1.rename(columns = {'_id':'task_id','type':'task_type','status':'task_status','store_id':'task_store_id'},inplace = True)
df_tasks1
```

```python
df_subtasks1 = df_subtasks[['_id','created_at','type','status','parent_task_id']]
df_subtasks1.rename(columns = {'_id':'subtask_id','type':'subtask_type','status':'subtask_status'},inplace = True)
df_subtasks1
```

```python
merged_df = pd.merge(df_tasks1,df_subtasks1, left_on = 'task_id',right_on = 'parent_task_id', how = 'inner')
merged_df
```

```python
sub_task_list = ['D2F_MOVEMENT','CC_SKIPPED_EXPIRY','CC_DEFECT','CC_SBC','SIMPLE_BIN_COUNT','MANUAL_UPLOAD','EXTERNAL_RTV','INTERNAL_RTV']
merged_df = merged_df[merged_df['subtask_type'].isin(sub_task_list)]
```

```python
merged_df['task_type'].unique()
```

```python
result = merged_df.groupby([ 'task_store_id']).agg(
    all_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x.nunique()),
    completed_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[merged_df['subtask_status'].isin(['COMPLETED', 'SKIPPED'])].nunique()),
    cc_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[merged_df['subtask_type'].isin(['CC_SKIPPED_EXPIRY', 'CC_DEFECT', 'CC_SBC'])].nunique()),
    cc_task_completed=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['subtask_type'].isin(['CC_SKIPPED_EXPIRY', 'CC_DEFECT', 'CC_SBC'])) & (merged_df['subtask_status'] == 'COMPLETED')].nunique()),
    d2f_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[merged_df['subtask_type'] == 'D2F_MOVEMENT'].nunique()),
    d2f_task_completed=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['subtask_type'] == 'D2F_MOVEMENT') & (merged_df['subtask_status'] == 'COMPLETED')].nunique()),
    expiry_removal_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['task_type'] == 'EXPIRY_ITEMS_REMOVAL') & (merged_df['creation_type'] == 'System')].nunique()),
    expiry_removal_task_completed=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['task_type'] == 'EXPIRY_ITEMS_REMOVAL') & (merged_df['creation_type'] == 'System') & (merged_df['subtask_status'] == 'COMPLETED')].nunique()),
    sbc_task=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['task_type'] == 'SIMPLE_BIN_COUNT') & (merged_df['creation_type'] == 'Manual')].nunique()),
    sbc_task_completed=pd.NamedAgg(column='subtask_id', aggfunc=lambda x: x[(merged_df['task_type'] == 'SIMPLE_BIN_COUNT') & (merged_df['creation_type'] == 'Manual') & (merged_df['subtask_status'] == 'COMPLETED')].nunique())
).reset_index()
```
